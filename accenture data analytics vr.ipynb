```python
m=[]
for i in df['Datetime'] :
    i=pd.to_datetime(i)
    m.append(i.month)
```

```python
m=pd.Series(m)
```

```python
df['Month']=m
```

```python
df.head()
```

```python
import numpy as np
```

```python
df['Count']=pd.Series(np.ones(len(df['Month'])))
```

```python
df3=df.groupby(by='Month').sum()
```

```python
df3=df3.sort_values(by='Count')
```
